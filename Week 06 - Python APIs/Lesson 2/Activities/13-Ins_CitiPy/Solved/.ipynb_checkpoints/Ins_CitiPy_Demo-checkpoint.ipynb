```python
# Dependencies
from citipy import citipy
import pandas as pd
import numpy as np
```

```python
# Some random coordinates
#coordinates = [(200, 200), (23, 200), (42, 100)]
lendf = 500
np.random.seed(10)
lats = np.random.uniform(-90,90, lendf).astype(float)
np.random.seed(50)
lons = np.random.uniform(-180,180, lendf).astype(float)
Lat_Lon =pd.DataFrame({"lats":lats,
            "lon":lons})

Lat_Lon


```

```python
 cities = []
for i, row in enumerate(Lat_Lon.values):
    lats = Lat_Lon.lats[i]
    lons =  Lat_Lon.lon[i]
    cities.append(citipy.nearest_city(lats,lons))
    print(cities[i].city_name)
    #now perform analysis on open/close based on date, etc..
    
    
```

```python
for city in cities:
    country_code = city.country_code
    name = city.city_name
    print(f"The country code of {name} is '{country_code}'.")
```
