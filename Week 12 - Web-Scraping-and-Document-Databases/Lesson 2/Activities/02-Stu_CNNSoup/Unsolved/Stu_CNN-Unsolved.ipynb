```python
# Dependencies
import os
from bs4 import BeautifulSoup as bs
```

```python
# Read HTML from file
filepath = os.path.join("..", "Resources", "template.html")
with open(filepath) as file:
    html = file.read()
    print(html)
```

```python
# Create a Beautiful Soup object
soup = bs(html)
type(soup)
print(soup.prettify())
```

```python
# Extract title text
soup.title
soup.title.text
```

```python
# Print all paragraph texts
para=soup.find_all('p')
for parag in para:
    print(parag.text.strip())
```

```python
# Print all ten headlines
tds=soup.find_all('td')
headlines=[]
for td in tds:
    if td.a:
        if td.a.text:
            headlines.append(td.a.text)
    
```

```python
# Print only the headlines
headlines[:10]
```

```python
for i in range(10):
    print(headlines[i])
```
